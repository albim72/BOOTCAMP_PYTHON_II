```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x = np.linspace(-2*np.pi,2*np.pi,100)
y = np.sinc(x)
```

```python
fig,ax = plt.subplots(ncols=2,figsize=(12,8))
ax[0].plot(x,y)
ax[0].set_title("pojedy≈Ñcza krzywa")
ax[1].plot(x,y)
ax[1].plot(x*2.0,y)
ax[1].set_title("Dwie krzywe")
plt.show()
```
