```python
from sympy import pi,Integer
class versin(Function):
    @classmethod
    def eval(cls,x):
        n = x/pi
        if isinstance(n,Integer):
            return 1-(-1)**2
```

```python
versin(15*pi)
```

```python
versin(2*pi)
```
